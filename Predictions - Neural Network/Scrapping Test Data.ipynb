```python
import pandas as pd
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from bs4 import BeautifulSoup
import time
```

```python
service = Service ('../Code/chromedriver-win64/chromedriver.exe')
```

```python
id = pd.read_excel("../Code/Unseen Dataset for NN.xlsx",usecols=['IMDb ID'])
id.head()
```

```python
def getpage (url):
    driver = webdriver.Chrome(service=service)
    driver.get(url)
    page_height = driver.execute_script("return document.body.scrollHeight;")
    print(page_height)
    driver.execute_script(f'window.scrollTo(0,{page_height}/3.3)')
    time.sleep(5)
    html = driver.page_source
    driver.quit()
    return html
```

```python
def extractingdetails(html):
    soup = BeautifulSoup(html,'lxml')
    genre = []
    title = soup.find('span', class_='hero__primary-text').text
    description = soup.find('span',{
        'class':'sc-2d37a7c7-2 PeLXr',
        'data-testid' :"plot-xl"
    }).text
    block = soup.find('section', {
    'cel_widget_id': "StaticFeature_Storyline",
    'class': "ipc-page-section ipc-page-section--base celwidget",
    'data-cel-widget': "StaticFeature_Storyline",
    'data-testid': "Storyline"
})
    if block:
        story = block.find('div',{'class' : 'ipc-html-content-inner-div'}).text
        story = description+story
        li_tags = block.find_all('a', {'class': 'ipc-metadata-list-item__list-content-item ipc-metadata-list-item__list-content-item--link'})
        for li_tag in li_tags:
            genre.append(li_tag.text)
    return title, genre, story
```

```python
data = []
for i, row in id.iterrows():
    imdb_id = row['IMDb ID']
    url = f"https://www.imdb.com/title/{imdb_id}/?ref_=fn_al_tt_1"
    try:
        print("Trying to get details...")
        html = getpage(url)
        title, genre, storyline = extractingdetails(html)
        data.append({'IMBb ID': imdb_id, 'Title': title, 'Storyline': storyline, 'Genres':genre})
        print(f'Details of {title} scrapped successfully!')
    except Exception as e:
        print("Error occurred!")
        print(f'Error: {e}')
```

```python
dataframe = pd.DataFrame(data)
dataframe
```

```python
dataframe.to_csv('test_data.csv')
```
