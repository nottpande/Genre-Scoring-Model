```python
import requests
import pandas as pd
import os
import re
from google.colab import files
```

```python
df1 = pd.read_csv('/content/textual_data_df.csv')
df1['Synopsis'] = df1['Synopsis'].str.replace('Plot Summary\n\n', '', regex=False)
df1
```

```python
df2 = pd.read_excel('/content/CC files for Fawesome movies.xlsx')
df2
```

#Getting The Subtitles

```python
dataset = pd.merge(df1, df2, left_on='node_id', right_on='nid', how='inner')
dataset
```

```python
def getsubtitles(row, download_dir='subtitles'):
    subtitle_url = row['cc_path']
    node_id = row['node_id']
    movie_name = row['title']
    # Create a safe filename
    filename = f"{movie_name}_{node_id}.srt".replace(" ", "_").replace("/", "_")
    dest_path = os.path.join(download_dir, filename)

    # Download the subtitle file
    response = requests.get(subtitle_url)
    if response.status_code == 200:
        with open(dest_path, 'wb') as file:
            file.write(response.content)

        # Attempt to read the subtitle content with utf-8 encoding
        try:
            with open(dest_path, 'r', encoding='utf-8') as file:
                return file.read()
        except UnicodeDecodeError:
            # If utf-8 fails, try a different encoding or handle errors
            try:
                with open(dest_path, 'r', encoding='latin1') as file:
                    return file.read()
            except UnicodeDecodeError:
                with open(dest_path, 'r', encoding='utf-8', errors='ignore') as file:
                    return file.read()
    else:
        print(f"Failed to download {movie_name}")
        return None
```

```python
os.makedirs('subtitles', exist_ok=True)
```

```python
def extract_text(subtitle):
    # Define a regular expression pattern to match timestamp lines
    timestamp_pattern = r'\d+\n\d{2}:\d{2}:\d{2},\d{3} --> \d{2}:\d{2}:\d{2},\d{3}\n'
    # Use re.sub to replace timestamp lines with an empty string
    text_only = re.sub(timestamp_pattern, '', subtitle)
    return text_only.strip()
```

```python
dataset['subtitles'] = dataset.apply(getsubtitles, axis=1)
```

```python
dataset = dataset.drop(['title','nid','cc_path'],axis=1)
new_order = ['node_id','IMDb Link','Title','subtitles','Synopsis','tags_string','extracted_genres']
dataset = dataset.reindex(columns=new_order)
dataset['subtitles'] = dataset['subtitles'].apply(extract_text)
dataset
```

#Adding the Summary Column

```python
gpt_summary = pd.read_csv('/content/gpt_summary.csv')
gpt_summary = gpt_summary.drop_duplicates('Title').reset_index().drop(['index'],axis=1)
gpt_summary
```

```python
gpt_summary.duplicated().sum()
```

```python
dataset_1 = pd.merge(dataset, gpt_summary, left_on='Title', right_on='Title', how='inner')
dataset_1 = dataset_1.drop(['Synopsis_x','tags_string_x','Unnamed: 0.1','Unnamed: 0','Synopsis_y','tags_string_y','tags','tags_truncated'],axis=1)
new_order = ['node_id','IMDb Link','Title','subtitles','summary','extracted_genres']
dataset_1 = dataset_1.reindex(columns=new_order)
dataset_1
```
